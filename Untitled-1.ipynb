```python
import pandas as pd
dfVehicles = pd.read_csv('vehicles.csv', header=0)
print(dfVehicles.describe)
```

```python
dfVehicles.drop(['barrels08', 'barrelsA08', 'charge120', 'charge240', 'city08',
       'city08U', 'cityA08', 'cityA08U', 'cityCD', 'cityE', 'cityUF', 'co2',
       'co2A', 'co2TailpipeAGpm', 'co2TailpipeGpm', 'comb08', 'comb08U',
       'combA08', 'combA08U', 'combE', 'combinedCD', 'combinedUF', 'engId', 'eng_dscr', 'feScore', 'fuelCost08',
       'fuelCostA08', 'fuelType1', 'ghgScore', 'ghgScoreA', 'highway08U', 'highwayA08', 'highwayA08U', 'highwayCD',
       'highwayE', 'highwayUF', 'hlv', 'hpv', 'id', 'lv2', 'lv4',
         'mpgData', 'phevBlended', 'pv2', 'pv4', 'range', 'rangeCity',
       'rangeCityA', 'rangeHwy', 'rangeHwyA', 'UCity', 'UCityA',
       'UHighway', 'UHighwayA', 'VClass', 'youSaveSpend', 'guzzler',
       'trans_dscr', 'tCharger', 'sCharger', 'atvType', 'fuelType2', 'rangeA',
       'evMotor', 'mfrCode', 'c240Dscr', 'charge240b', 'c240bDscr',
       'createdOn', 'modifiedOn', 'startStop', 'phevCity', 'phevHwy',
       'phevComb'], axis = 1, inplace=True)
```

```python
print(dfVehicles)
```

```python
dfVehicles.rename(columns = {'cylinders':'cilindros', 'displ':'consumo_litros', 'drive':'eje_motriz', 'fuelType':'tipo_combustible', 
'highway08':'mpg', 'make':'marca', 'model':'modelo', 'trany':'transmision', 'year':'anio' }, inplace = True)
print(dfVehicles)
```

```python
dfVehicles = dfVehicles[(dfVehicles.marca.isin(['Audi','BMW','Porsche','Volvo','Mercedes-Benz']) & 
(dfVehicles.tipo_combustible.isin(['Premium','Electricity'])) &
(dfVehicles['anio'].between(2020,2023)))]
print(dfVehicles)
```

```python
dfVehicles.sort_values(by='anio', ascending = True, inplace=True)
print(dfVehicles)
```

```python
dfPremium = dfVehicles.query("tipo_combustible == 'Premium'")
dfElectricity = dfVehicles.query("tipo_combustible == 'Electricity'")

dfPremium = dfPremium.marca.value_counts().reset_index(name="count")
dfElectricity = dfElectricity.marca.value_counts().reset_index(name="count")
dfPremium.sort_values(by='index', ascending = True, inplace=True)
dfElectricity.sort_values(by='index', ascending = True, inplace=True)
print(dfPremium)
print(dfElectricity)
```

```python
import matplotlib.pyplot as plt
plt.subplot()
plt.bar(dfPremium['index'], dfPremium['count'], width = 0.75, label='Premium')
plt.bar(dfElectricity['index'], dfElectricity['count'],bottom=dfPremium['count'], width = 0.75, label='Electricity')
plt.legend(['Premium','Electricity'])

plt.show()
```
